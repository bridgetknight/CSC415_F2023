```python
# imports
import pandas as pd
```

```python
# function to sort books using selection sort
def sort_years(file):
    
    # Read in the data and process it
    print("Reading file...")
    books = pd.read_csv(file, engine="python")
    years = (books["Year-Of-Publication"]._convert(numeric=True).dropna()).unique()

    # Sort years
    print("Sorting years...")
    n = len(years)
    for i in range(n):
        min_index = i
        for j in range(min_index+1, n):
            if(years[j] < years[min_index]):
                min_index = j
        (years[i], years[min_index]) = (years[min_index], years[i])
    

    
    return years
            
    
```

```python
# function to sort books using selection sort
def sort_arrivals(file):
    
    # Read in the data and process it
    print("Reading file...")
    books = pd.read_csv(file, engine="python")

    # Sort based on publication year
    print("Sorting...")
    n = len(books)
    for i in range(n):
        min_index = i
        print(books.iloc[i])
        for j in range(i+1, n):
            if(books.iloc[j]["Year-Of-Publication"] < books.iloc[min_index]["Year-Of-Publication"]):
                min_index = j
        (books.iloc[i], books.iloc[min_index]) = (books.iloc[min_index].copy(), books.iloc[i].copy())
        
    print("Done!")
    
    # Return 50 most recent books
    new_arrivals = books.head(50)
    
    return new_arrivals
            
    
```
